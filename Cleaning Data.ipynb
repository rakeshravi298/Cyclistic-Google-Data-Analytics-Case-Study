```python
#to_epoch

def to_epoch(s):
    year = int(s[0:4])
    month = int(s[5:7])
    date = int(s[8:10])
    hour = int(s[11:13])
    minute = int(s[14:16])
    second = int(s[17:19])
    
    return dt.datetime(year,month,date,hour,minute,second).timestamp()
    
    
```

```python
def month(s):
    return int(s[5:7])

def date(s):
    return int(s[8:10])
```

```python
min_epoch = min(df["start_time_epoch"])
print(min_epoch)
def epoch_time_minus(s):
    return s-min_epoch
    
```

```python
df.loc[df["start_station_id"].isna() , "start_station_id"]= 0
df.loc[df["end_station_id"].isna() , "end_station_id"]= 0
df.loc[df["end_station_name"].isna() , "end_station_name"]= "unavailable"
df.loc[df["start_station_name"].isna() , "start_station_name"]= "unavailable"
df.loc[df["start_lat"].isna() , "start_lat"]= 0
df.loc[df["start_lng"].isna() , "start_lng"]= 0
df.loc[df["end_lat"].isna() , "end_lat"]= 0
df.loc[df["end_lng"].isna() , "end_lng"]= 0
```

```python
df["start_time_epoch"] = df["started_at"].apply(to_epoch)
df["end_time_epoch"] = df["ended_at"].apply(to_epoch)
```

```python
df["month"] = df["started_at"].apply(month)
df["date"] = df["started_at"].apply(date)
```

```python
df = df.sort_values(by=['start_time_epoch'])
```

```python
df["start_epoch"] = df["start_time_epoch"].apply(epoch_time_minus)
df["end_epoch"] = df["end_time_epoch"].apply(epoch_time_minus)
```

```python
df["ride_time"] = round((df["end_epoch"] - df["start_epoch"])/60 , 3)
```
